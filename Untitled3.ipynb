```python
from conf import *
from subprocess import check_output
import subprocess
from pathlib import Path
import csv
import time
import pandas as pd


start = time.time()

######################################################
#                      SLEUTH                        #
######################################################

fn = ''

# Reading Encoding list
r = open(f'encoding_lists/encodingList_TTR-{trTimeThresh}_{CITY}{fn}', "r")
data = r.read()
encodingList = data.split('\n')
r.close()

encodingList = filter(None, encodingList) 

dst_path = Path('kdd_lstw_extraction')
if not dst_path.is_dir():
    dst_path.mkdir(parents=True)


def codeToLabel(code):
    codeToLabel = {}
    codeToLabel['-1'] = '-1'
    codeToLabel['1'] = 'Snow'
    codeToLabel['2'] = 'Rain'
    codeToLabel['3'] = 'Construction'
    codeToLabel['4'] = 'Congestion'
    codeToLabel['5'] = 'Event'
    codeToLabel['6'] = 'Fog'
    codeToLabel['7'] = 'Lane-Blocked'
    codeToLabel['8'] = 'Cold'
    codeToLabel['9'] = 'Other'
    codeToLabel['10'] = 'Storm'
    codeToLabel['11'] = 'Broken-Vehicle'
    codeToLabel['12'] = 'Incident-Weather'
    codeToLabel['13'] = 'Precipitation-UNK'
    codeToLabel['14'] = 'Hail-Other'
    codeToLabel['15'] = 'Incident-Other'
    codeToLabel['16'] = 'Incident-Flow'
    codeToLabel['17'] = 'Accident'
    
    return codeToLabel[code]

def sleuth(z, encodings, i_e = 'E'):
    total=passed = 0
    # make temporary db for encodings
    input_name = 'input_' + file_name + '.db'
    w = open(dst_path / input_name, 'w')
    lg = 0
    for e in encodings: 
        l = int(e.split(' ')[2])        
        if l<max_tree_length:       
            w.write(e + '\n')
            lg+=1           
    w.close()
    
    total += len(encodings)
    passed += lg
    min_sup = min_sup_fixed

    # run sleuth algorithm to find frequent induced/embedded unordered tree patterns
    print("\nStarting SLEUTH...\n")
    out = check_output(['./sleuth', '-i', (dst_path / input_name).as_posix(), '-S', str(min_sup), '-o']).decode('utf-8')
    print(out)
    print("SLEUTH has finished\n")
    w = open(dst_path / file_name, 'a')
    out = out.split('F')[0].split(')')[1].split('\n')
    for p in out:
        if len(p)==0: continue
        sp = int(p.split(' - ')[1])*1.0
        p = p.split('- ')[0]
        patterns = p.split(' ')
        patterns = filter(None, patterns)
        patterns_enc = list(map(codeToLabel, patterns))
        patterns_enc2 = ' âžœ '.join(patterns_enc)
        w.write(z + ',' + patterns_enc2 + ',' + str(round(sp/lg, 3)) + ',' + str(sp) + '\n')
        
    w.close()
    return total, passed


# the output file for frequent tree patterns
file_name = 'frequent_trees_City_MSF-{}_MTL-{}_TTR-{}_{}.csv'.format(min_sup_fixed, max_tree_length, trTimeThresh, CITY)

w = open(dst_path / file_name, 'w')
w.write('City,Pattern,Support,SupportABS\n')
w.close()

print ('CITY:', CITY)
print ('max_tree_length:', max_tree_length)
print ('min_sup_fixed:', min_sup_fixed)
print ('trTimeThresh:', trTimeThresh)

writer = open(dst_path / f'State_TreeCount_CityBased_{CITY}_{min_sup_fixed}_{max_tree_length}_{trTimeThresh}.csv', 'w')
writer.write('State,Count\n')
state_total_pattern = {}

encodingSet = []
for enc in encodingList:
    _parts = enc.split(' ')            
    enc = str(len(encodingSet)+1) + ' ' + str(len(encodingSet)+1)
    for i in range(2,len(_parts)): enc += ' ' + _parts[i]
    encodingSet.append(enc)
Total, Passed = sleuth(CITY, encodingSet, 'E')
_state = CITY
if _state in state_total_pattern: state_total_pattern[_state] = state_total_pattern[_state] + Passed
else: state_total_pattern[_state] = Passed

print('Total Trees: %d, Total Passed the Criteria: %d, Pass Ratio %.2f' %(Total, Passed, (float(Passed)/Total)*100))    
for st in state_total_pattern:
    writer.write(st + ',' + str(state_total_pattern[st]) + '\n')
writer.close()

print("\nElapsed Time: %.2fs" % (time.time()-start))
```
